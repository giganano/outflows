# The ISM radial velocity profile from our GCE model inputs.

The ISM radial velocity profile is sensitive to the accretion rate of the Galaxy due to angular momentum transport. Based on this connection, how is the radial velocity profile related to our GCE model parameters?

For a patch of ISM gas accreting material, the change in the total angular momentum over the course of one timestep is given by

$$
L_g = M_g v_{\phi,g} R_\text{gal} \rightarrow
M_g v_{\phi,g} R_\text{gal} + \dot{M}_\text{in} \delta t v_{\phi,\text{in}} R_\text{gal},
$$

where $M_g$ is the mass of the fluid element, $\dot{M}_\text{in}$ is its accretion rate, $v_\phi$ refers to the circular velocity of the ISM and the accreting material, and $R_\text{gal}$ is Galactocentric radius. In general, $v_{\phi,\text{in}} < v_{\phi,g}$. The lower specific angular momentum of the material is more or less the reason why it is accreting in the first place. As a result, the ISM angular momentum is diluted, causing it to fall to a smaller radius at take on the circular velocity at that radius. The new angular momentum,

$$
L_g' = \left(M_g + \dot{M}_\text{in} \delta t\right)
\left(v_{\phi,g} + \sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i \right)
\left(R_\text{gal} + \delta R_\text{gal}\right),
$$

must be equivalent to the initial angular momentum $L_g$. Therefore

$$
\begin{align}
M_g v_{\phi,g} R_\text{gal} + \dot{M}_\text{in} \delta t v_{\phi,\text{in}} R_\text{gal} &=
\left(M_g + \dot{M}_\text{in} \delta t\right)
\left(v_{\phi,g} + \sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i \right)
\left(R_\text{gal} + \delta R_\text{gal}\right)
\\
%&= M_g
%\left(v_{\phi,g} + \sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i \right)
%\left(R_\text{gal} + \delta R_\text{gal}\right) +
%\dot{M}_\text{in} \delta t
%\left(v_{\phi,g} + \sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i \right)
%\left(R_\text{gal} + \delta R_\text{gal}\right)
%\\
\dot{M}_\text{in} R_\text{gal} \left[
v_{\phi,\text{in}} - v_{\phi,g} - v_{\phi,g} \frac{\delta R_\text{gal}}{R_\text{gal}} -
\sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i -
\frac{\delta R_\text{gal}}{R_\text{gal}}
\sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^{i}
\right] &=
M_g R_\text{gal} \left[
v_{\phi,g} \frac{\delta R_\text{gal}}{R_\text{gal} \delta t} +
\frac{1}{\delta t} \sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i +
\frac{\delta R_\text{gal}}{R_\text{gal} \delta t}
\sum_{i = 1}^\infty \frac{\partial^i v_{\phi,g}}{\partial R_\text{gal}^i} \delta R_\text{gal}^i
\right]
\end{align}
$$

In taking the limit at $\delta R, \delta t \rightarrow 0$, the left-hand side simplifies to $\dot{M}_\text{in} R_\text{gal} \left(v_{\phi,\text{in}} - v_{\phi,g}\right)$. On the right-hand side, both summations approach $\infty$ as $\delta t \rightarrow 0$ but approach $0$ as $\delta R_\text{gal} \rightarrow 0$. Their limits are therefore undefined, so the only physical value we can attach to them in the differential case is zero. Therefore

$$
\dot{M}_\text{in} R_\text{gal} \left(v_{\phi,\text{in}} - v_{\phi,g}\right) = M_g v_{\phi,g} \langle v_r \rangle,
$$

where $\langle v_r \rangle = dR_\text{gal} / dt$ is the mean radial velocity of the gas. The proper radial gas velocity to assign to a given ISM fluid element is then given by

$$
\langle v_r \rangle = \frac{\dot{M}_\text{in}}{M_g} R_\text{gal} \left(\beta_\phi - 1\right),
$$

where $\beta_\phi \equiv v_{\phi,\text{in}} / v_{\phi,g}$ is the ratio of circular velocities between the accreting material and the disk gas.

We can now relate the radial gas velocity to our input GCE parameters by substituting in for the infall rate and gas supply. The rate of change in the gas supply:

$$
\begin{align}
\dot{M}_g &= \dot{M}_\text{in} - \dot{M}_\star \left(1 + \eta - r - \gamma_\text{flow}\right)
\\
\implies \dot{M}_\text{in} &= \dot{M}_g + \dot{M}_\star \left(1 + \eta - r - \gamma_\text{flow}\right)
\\
\implies \frac{\dot{M}_\text{in}}{M_g} &=
\frac{\dot{M}_g}{M_g} + \frac{1 + \eta - r}{\tau_\star} +
\langle v_r \rangle \left[
\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}
\right] +
\frac{\partial \langle v_r \rangle}{\partial R_\text{gal}}.
\end{align}
$$

Plugging the above into the expression above that:

$$
\langle v_r \rangle = \left[
\frac{\dot{M}_g}{M_g} + \frac{1 + \eta - r}{\tau_\star} +
\langle v_r \rangle
\left(\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}\right) +
\frac{\partial \langle v_r \rangle}{\partial R_\text{gal}}
\right] R_\text{gal} \left(\beta_\phi - 1\right)
$$

$$
\begin{align}
\implies \frac{\partial \langle v_r \rangle}{\partial R_\text{gal}} &=
\frac{\langle v_r \rangle}{R_\text{gal} \left(\beta_\phi - 1\right)} -
\frac{\dot{M}_g}{M_g} - \frac{1 + \eta - r}{\tau_\star} -
\langle v_r \rangle \left(\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}\right)
\\
&= -\frac{\dot{M}_g}{M_g} - \frac{1 + \eta - r}{\tau_\star} -
\langle v_r \rangle \left[
\frac{1}{R_\text{gal}}\left(1 - \frac{1}{\beta_\phi - 1}\right) +
\frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}\right]
\\
&= -\frac{\dot{M}_g}{M_g} - \frac{1 + \eta - r}{\tau_\star} -
\langle v_r \rangle \left[
\frac{1}{R_\text{gal}}\left(\frac{\beta_\phi - 2}{\beta_\phi - 1}\right) +
\frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}\right].
\end{align}
$$

Now transforming from mass $M_g$ to surface density $\Sigma_g$ results in the following expression for the velocity profile $\partial \langle v_r \rangle / \partial R_\text{gal}$:

$$
\frac{\partial \langle v_r \rangle}{\partial R_\text{gal}} =
-\frac{\dot{\Sigma}_g}{N \Sigma_g} - \frac{1 + \eta - r}{\tau_\star} -
\langle v_r \rangle \left[
\frac{1}{R_\text{gal}}\left(\frac{\beta_\phi - 2}{\beta_\phi - 1}\right) +
\frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}\right].
$$

The earlier expression $\langle v_r \rangle (\dot{M}_\text{in} / M_g) R_\text{gal} (\beta_\phi - 1)$ indicates that $\langle v_r \rangle = 0$ at $R_\text{gal} = 0$ (the infall rate per unit gas supply will approach some finite value in the limit that $R_\text{gal} \rightarrow 0$). Therefore, the above expression constitutes an initial value problem that we can solve at each timestep to obtain the full solution to the radial gas velocity as a function of radius and time.

After plugging this expression back into $\mu_\text{flow}$:

$$
\begin{align}
\mu_\text{flow} &\equiv -\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}} +
\frac{\partial \ln \langle v_r \rangle}{\partial R_\text{gal}} +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right]
\\
&= -\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}} +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right] - \tau_\star \frac{\partial \langle v_r \rangle}{\partial R_\text{gal}}
\\
&= -\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} + \frac{\partial \ln \Sigma_g}{\partial R_\text{gal}} +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right] +
\tau_\star \frac{\dot{\Sigma}_g}{\Sigma_g} +
1 + \eta - r +
\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} \left(\frac{\beta_\phi - 2}{\beta_\phi - 1}\right) +
\frac{\partial \ln \Sigma_g}{\partial R_\text{gal}}
\right]
\\
&= -\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} \left(1 - \frac{\beta_\phi - 2}{\beta_\phi - 1}\right) +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right] +
\tau_\star \frac{\dot{\Sigma}_g}{\Sigma_g} +
1 + \eta - r
\\
&= -\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}} \left(\frac{1}{\beta_\phi - 1}\right) +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right] +
\tau_\star \frac{\dot{\Sigma}_g}{\Sigma_g} +
1 + \eta - r,
\end{align}
$$

we arrive at the following relation

$$
1 + \eta - r - \mu_\text{flow} =
\tau_\star \langle v_r \rangle \left[
\frac{1}{R_\text{gal}}
\left(\frac{1}{\beta_\phi - 1}\right) +
\frac{\partial \ln Z}{\partial R_\text{gal}}
\right] -
\tau_\star \frac{\dot{\Sigma}_g}{\Sigma_g},
$$

which has an interesting implication when inserted into an expression for the equilibrium abundance. For an SFH in which the late time behavior is an exponential decline, $Z_\text{O,eq}$ is given by

$$
Z_\text{O,eq} = \frac{y_\text{O}^\text{CC}}{1 + \eta - r - \mu_\text{flow} - \tau_\star / \tau_\text{sfh}},
$$

where $y_\text{O}^\text{CC}$ is the population-averaged O yield from massive stars and $\tau_\text{sfh}$ is the e-folding timescale of the late-time SFH. If we adopt a power-law Kennicutt-Schmidt relation $\dot{\Sigma}_\star \propto \Sigma_g^N$, then $\dot{\Sigma}_g / \Sigma_g = \ddot{\Sigma}_\star / N\dot{\Sigma}_\star \rightarrow -1 / N\tau_\text{sfh}$.

Our expression for the equilibrium abundance now becomes

$$
\begin{align}
Z_\text{O,eq} &= y_\text{O}^\text{CC} \left[
\tau_\star \langle v_r \rangle \left(\frac{1}{R_\text{gal} (\beta_\phi - 1)} +
\frac{\partial \ln Z}{\partial R_\text{gal}}\right) +
\frac{\tau_\star}{N \tau_\text{sfh}} -
\frac{\tau_\star}{\tau_\text{sfh}}
\right]^{-1}
\\
&= y_\text{O}^\text{CC} \left[
\tau_\star \langle v_r \rangle \left(\frac{1}{R_\text{gal} (\beta_\phi - 1)} +
\frac{\partial \ln Z}{\partial R_\text{gal}}\right) +
\frac{\tau_\star}{\tau_\text{sfh}}
\left(\frac{1}{N} - 1\right)
\right]^{-1}
\end{align}
$$

```python
def plot_benchmark_gradient(ax, filename = "./empirical/skewnormfits-binned-by-radius.out", element = "O",
    **kwargs):
    raw = np.genfromtxt(filename)
    radii = [(a + b) / 2 for a, b in zip (raw[:, 0], raw[:, 1])]
    if element.lower() == "o":
        mode = raw[:, 2]
    elif element.lower() == "fe":
        mode = raw[:, 6]
    else:
        raise ValueError("Don't have gradient measurements for element: %s" % (element))
    ax.scatter(radii, mode, **kwargs)
    
def boxcarsmoothtrend(xvals, yvals, window = 5):
    newyvals = len(yvals) * [0.]
    for i in range(len(xvals)):
        lower = max(i - window, 0)
        upper = min(i + window, len(xvals) - 1)
        newyvals[i] = np.mean(yvals[lower:upper])
    return newyvals

def radial_velocity_profile(output, lookback, zone_width = 0.1, disk_lifetime = 13.2):
    raw = np.genfromtxt("%s_gasvelocities.out" % (output.name))
    n_zones = 0
    while raw[n_zones][1] == 0: n_zones += 1
    time = disk_lifetime - lookback
    diff = [abs(_ - time) for _ in raw[:, 1]]
    start = diff.index(min(diff))
    radii = raw[:, 0][start:start + n_zones]
    vr = raw[:, 2][start:start + n_zones]
    return [radii, vr]

def radial_velocity_vs_time(output, zone_index):
    raw = np.genfromtxt("%s_gasvelocities.out" % (output.name))
    n_zones = 0
    while raw[n_zones][1] == 0: n_zones += 1
    raw = raw[zone_index::n_zones]
    time = raw[:, 1]
    vr = raw[:, 2]
    return [time, vr]

def mu(output, lookback, element = "O", zone_width = 0.1, disk_lifetime = 13.2, **kwargs):
    _, vr = radial_velocity_profile(output, lookback, zone_width = zone_width, disk_lifetime = disk_lifetime)
    radii = np.arange(0, 20, zone_width)
    mu_x = len(radii) * [0.]
    diff = [abs(lookback - _) for _ in output.zones["zone0"].history["lookback"]]
    idx = diff.index(min(diff))
    for i in range(len(radii) - 2):
        zone = output.zones["zone%d" % (i)]
        neighbor = output.zones["zone%d" % (i)]
        area = np.pi * (radii[i + 1]**2 - radii[i]**2)
        area_n = np.pi * (radii[i + 2]**2 - radii[i + 1]**2)
        mgas = zone.history["mgas"][idx]
        sfr = zone.history["sfr"][idx]
        tau_star = 1.e-9 * mgas / sfr if sfr else float("inf")
        sigma_g = zone.history["mgas"][idx] / area
        sigma_g_n = neighbor.history["mgas"][idx] / area_n
        if element is not None:
            z = zone.history["z(%s)" % (element)][idx]
            z_n = neighbor.history["z(%s)" % (element)][idx]
        else: pass
        mu_x[i] = 1 / radii[i] if radii[i] else float("inf")
        mu_x[i] += (sigma_g_n - sigma_g) / (sigma_g * zone_width)
        mu_x[i] += (vr[i + 1] - vr[i]) / (vr[i] * zone_width) if vr[i] else float("inf")
        if element is not None:
            mu_x[i] += (z_n - z) / (z * zone_width)
        else: pass
        mu_x[i] *= -tau_star * vr[i]
    return [radii, mu_x]

def mu_vs_time(output, zone_index, element = "O", zone_width = 0.1, disk_lifetime = 13.2, **kwargs):
    _, vr = radial_velocity_vs_time(output, zone_index)
    _, vr_n = radial_velocity_vs_time(output, zone_index + 1)
    radii = np.arange(0, 20, zone_width)
    zone = output.zones["zone%d" % (zone_index)]
    neighbor = output.zones["zone%d" % (zone_index + 1)]
    area = np.pi * (radii[zone_index + 1]**2 - radii[zone_index]**2)
    area_n = np.pi * (radii[zone_index + 2]**2 - radii[zone_index + 1]**2)
    time = zone.history["time"]
    mu_x = len(time) * [0.]
    for i in range(len(time)):
        sigma_sfr = zone.history["sfr"][i] / area
        sigma_g = zone.history["mgas"][i] / area
        sigma_g_n = neighbor.history["mgas"][i] / area_n
        tau_star = 1.e-9 * sigma_g / sigma_sfr if sigma_sfr else float("inf")
        if element is not None:
            z = zone.history["z(%s)" % (element)][i]
            z_n = neighbor.history["z(%s)" % (element)][i]
        else: pass
        mu_x[i] = 1 / radii[zone_index] if radii[zone_index] else float("inf")
        mu_x[i] += (sigma_g_n - sigma_g) / (sigma_g * zone_width)
        mu_x[i] += (vr_n[i] - vr[i]) / (vr[i] * zone_width) if vr[i] else float("inf")
        if element is not None:
            mu_x[i] += (z_n - z) / (z * zone_width) if z else float("inf")
        else: pass
        mu_x[i] *= -tau_star * vr[i]
    return [time, mu_x]

def gamma_coefficients_vs_time(output, zone_index, zone_width = 0.1, betaphi_out = 0, betaphi_in = 0.7,
    disk_lifetime = 13.2, element = "O"):
    zone = output.zones["zone%d" % (zone_index)]
    neighbor = output.zones["zone%d" % (zone_index + 1)]
    time = zone.history["time"]
    # gamma_eta = len(time) * [0.]
    # gamma_star = len(time) * [0.]
    # gamma_mu = len(time) * [0.]
    gamma_disk = len(time) * [0.]
    gamma_ej = len(time) * [0.]
    gamma_xi = len(time) * [0.]
    _, vr = radial_velocity_vs_time(output, zone_index)
#     _, mu_g = mu_vs_time(output, zone_index, element = element, zone_width = zone_width,
#         disk_lifetime = disk_lifetime)
    for i in range(len(time)):
#         sys.stdout.write("\r a")
#         _, mu_g = mu(output, disk_lifetime - time[i], element = None)
#         sys.stdout.write("\r b")
#         sys.stdout.flush()
        if zone.history["sfr"][i]:
            tau_star = zone.history["mgas"][i] / zone.history["sfr"][i] * 1e-9 # yr^-1 -> Gyr^-1
        else:
            tau_star = float("inf")
        # gamma_star[i] = -tau_star * vr[i] / (zone_width * zone_index * (1 - betaphi_in))
        gamma_disk[i] = -tau_star * vr[i] / (zone_width * zone_index * (1 - betaphi_in))
        # gamma_eta[i] = zone.history["eta_0"][i] * (1 - betaphi_out) / (1 - betaphi_in)
        gamma_ej[i] = zone.history["eta_0"][i] * (1 - betaphi_out) / (1 - betaphi_in)
        dzdr = (neighbor.history["z(%s)" % (element)][i] - zone.history["z(%s)" % (element)][i]) / zone_width
        if zone.history["z(%s)" % (element)][i]:
            # gamma_mu[i] = tau_star * vr[i] * dzdr / zone.history["z(%s)" % (element)][i]
            gamma_xi[i] = tau_star * vr[i] * dzdr / zone.history["z(%s)" % (element)][i]
        else:
            # gamma_mu[i] = float("inf")
            gamma_xi[i] = float("inf")
        if zone.history["z(%s)" % (element)][i]:
            xi_out = zone.history["z_out(%s)" % (element)][i] / zone.history["z(%s)" % (element)][i]
        else:
            xi_out = float("nan")
        gamma_xi[i] -= (1 - xi_out) * zone.history["eta_0"][i]
#     sys.stdout.write("\n")
    # return [time, gamma_eta, gamma_star, gamma_mu]
    return [time, gamma_disk, gamma_ej, gamma_xi]

def gamma_coefficients_vs_radius(output, lookback, zone_width = 0.1, betaphi_out = 0, betaphi_in = 0.7,
    disk_lifetime = 13.2, element = "O"):
    radii, vr = radial_velocity_profile(output, lookback, zone_width = zone_width, disk_lifetime = disk_lifetime)
    diff = [abs(_ - lookback) for _ in output.zones["zone0"].history["lookback"]]
    idx = diff.index(min(diff))
    # gamma_eta = len(radii) * [0.]
    # gamma_star = len(radii) * [0.]
    # gamma_mu = len(radii) * [0.]
    gamma_disk = len(radii) * [0.]
    gamma_eta = len(radii) * [0.]
    gamma_mu = len(radii) * [0.]
#     _, mu_g = mu(output, lookback, element = None)
    for i in range(len(radii)):
        zone = output.zones["zone%d" % (i)]
        # xi_out = zone.history["z_out(%s)" % (element)][idx] / zone.history["z(%s)" % (element)][idx]
        # gamma_eta[i] = zone.history["eta_0"][idx] * (1 - betaphi_out) / (1 - betaphi_in)
        # gamma_eta[i] += (xi_out - 1) * zone.history["eta_0"][idx]
        if zone.history["sfr"][idx]:
            tau_star = zone.history["mgas"][idx] / zone.history["sfr"][idx] * 1e-9
        else:
            tau_star = float("inf")
        if radii[i]:
            # gamma_star[i] = -tau_star * vr[i] / (radii[i] * (1 - betaphi_in))
            gamma_disk[i] = -tau_star * vr[i] / (radii[i] * (1 - betaphi_in))
        else:
            # gamma_star[i] = -float("inf")
            gamma_disk[i] = -float("inf")
        if zone.history["z(%s)" % (element)][idx]:
            xi_out = zone.history["z_out(%s)" % (element)][idx] / zone.history["z(%s)" % (element)][idx]
        else:
            xi_out = float("nan")
        gamma_eta[i] = zone.history["eta_0"][idx] * ((1 - betaphi_out) / (1 - betaphi_in) + xi_out - 1)
        if i != len(radii) - 1:
            neighbor = output.zones["zone%d" % (i + 1)]
            dzdr = (neighbor.history["z(%s)" % (element)][idx] - zone.history["z(%s)" % (element)][idx]) / zone_width
            gamma_mu[i] = tau_star * vr[i] * dzdr / zone.history["z(%s)" % (element)][idx]
            # gamma_mu[i] = tau_star * vr[i] * dzdr / zone.history["z(%s)" % (element)][idx]
        else:
            # gamma_mu[i] = gamma_mu[i - 1]
            gamma_mu[i] = float("inf")
    # return [radii, gamma_eta, gamma_star, gamma_mu]
    return [radii, gamma_disk, gamma_eta, gamma_mu]

def zeq_vs_time(output, zone_index, element = "O", zone_width = 0.1, yocc = 0.00713, **kwargs):
    time, gamma_disk, gamma_eta, gamma_mu = gamma_coefficients_vs_time(output, zone_index,
        zone_width = zone_width, element = element, **kwargs)
    gamma_tot = [a + b + c for a, b, c in zip(gamma_disk, gamma_eta, gamma_mu)]
    zeq = len(time) * [0.]
    zone = output.zones["zone%d" % (zone_index)]
    for i in range(len(time)):
        z_in = zone.history["z_in(%s)" % (element)][i]
        zeq[i] = yocc / gamma_tot[i] + z_in * (1 - gamma_mu[i] / gamma_tot[i])
#         gamma_tot = gamma_disk[i] + gamma_ej[i] + gamma_xi[i]
#         eta = zone.history["eta_0"][i]
#         z_ism = zone.history["z(%s)" % (element)][i]
#         z_in = zone.history["z_in(%s)" % (element)][i]
#         z_out = zone.history["z_out(%s)" % (element)][i]
#         xi_out = z_out / z_ism if z_ism else float("nan")
#         x = yocc + z_in * (gamma_eta[i] + gamma_star[i] - (xi_out - 1) * eta)
#         zeq[i] = x / gamma_tot
    return [time, zeq]
```

```python
def molecular_tau_star(time):
    return 2 * ((0.5 + time) / 13.7)**0.5

def eta(r, yieldsolar = 3, slope = -0.062):
    etasun = yieldsolar - 0.6
    return etasun * np.exp(-slope * np.log(10) * (r - 8))

def eta_zero(r):
    return 0

def sfe(time, sigma_sfr):
    # print(sigma_sfr / 1e9)
    # if sigma_sfr == 0: return float("nan")
    mol = molecular_tau_star(time)
    N = plaw_index(time, sigma_sfr)
    return mol * (sigma_sfr * mol / 1e8)**(1 / N - 1)

def plaw_index(time, sigma_sfr):
    mol = molecular_tau_star(time)
    if sigma_sfr > 1e8 / mol:
        return 1
    else:
        return 1.5
```

```python
def boxcarsmoothtrend(times, values, window = 0.25):
    smoothed = []
    for i in range(len(times)):
        idxmin = None
        idxmax = None
        for j in range(len(times)):
            if idxmin is None and times[j] >= times[i] - window: idxmin = j
            if idxmax is None and times[j] >= times[i] + window: idxmax = j - 1
            if idxmin is not None and idxmax is not None: break
        if idxmax is None: idxmax = i
        smoothed.append(np.mean(values[idxmin:idxmax + 1]))
    return smoothed
```

```python
def radial_velocity_profile(evolution = "insideout", zone_width = 0.1, betaphi_out = 0, betaphi_in = 0.7):
    radii, vr = radial_gas_velocity_profile(
        star_formation_history(spec = evolution, zone_width = zone_width),
        inputs.eta_function,
        inputs.sfe_function,
        inputs.plaw_index,
        lambda r, t: betaphi_in,
        lambda r, t: betaphi_out)(13.2)
    stop = 0
    for i in range(len(radii)):
        if radii[i] > 15:
            stop = i
            break
    return [radii[:stop], vr[:stop]]
    # radii = np.arange(0, 15, zone_width)
    # vr = len(radii) * [0.]
    # for i in range(len(vr)):
    #     vr[i] = out.zones["zone%d" % (i)].history["ofr"][200] * (1 - betaphi_out) * 1e9
    #     vr[i] -= out.zones["zone%d" % (i)].history["ifr"][200] * (1 - betaphi_in) * 1e9
    #     vr[i] *= radii[i] / out.zones["zone%d" % (i)].history["mgas"][200]
    # return [radii, vr]

def gamma_coefficients(out, evolution = "insideout", zone_width = 0.1, betaphi_out = 0, betaphi_in = 0.7):
    print("gamma_coefficients: betaphi_out = %.2f" % (betaphi_out))
    radii, vr = radial_velocity_profile(evolution = evolution, zone_width = zone_width,
        betaphi_out = betaphi_out, betaphi_in = betaphi_in)
    gamma_eta = len(radii) * [0.]
    for i in range(len(gamma_eta)):
        gamma_eta[i] = out.zones["zone%d" % (i)].history["eta_0"][-1] * (1 - betaphi_out) / (1 - betaphi_in)
    gamma_star = len(radii) * [0.]
    for i in range(len(radii)):
        zone = out.zones["zone%d" % (i)]
        if zone.history["sfr"][-1]:
            tau_star = zone.history["mgas"][-1] / zone.history["sfr"][-1] * 1e-9
        else:
            tau_star = float("nan")
        if radii[i]:
            gamma_star[i] = -tau_star * vr[i] / (radii[i] * (1 - betaphi_in))
        else:
            gamma_star[i] = float("nan")
    gamma_mu = len(radii) * [0.]
    for i in range(len(gamma_mu)):
        zone = out.zones["zone%d" % (i)]
        neighbor = out.zones["zone%d" % (i + 1)]
        if zone.history["sfr"][-1]:
            tau_star = zone.history["mgas"][-1] / zone.history["sfr"][-1] * 1e-9
        else:
            tau_star = float("nan")
        dzdr = (neighbor.history["z(o)"][-1] - zone.history["z(o)"][-1]) / zone_width
        gamma_mu[i] = tau_star * vr[i] * dzdr / zone.history["z(o)"][-1]
    return [radii, gamma_eta, gamma_star, gamma_mu]

def zdot(out, zone_width = 0.1, **kwargs):
    # print("gamma_coefficients: betaphi_out = %.2f" % (kwargs["betaphi_out"]))
    radii, zdot_eta, zdot_star, zdot_mu = gamma_coefficients(out, zone_width = zone_width, **kwargs)
    for i in range(len(radii)):
        zone = out.zones["zone%d" % (i)]
        tau_star = zone.history["mgas"][-1] / zone.history["sfr"][-1] * 1e-9
        zdot_eta[i] *= -zone.history["z(o)"][-1] / tau_star
        zdot_star[i] *= -zone.history["z(o)"][-1] / tau_star
        zdot_mu[i] *= -zone.history["z(o)"][-1] / tau_star
    return [radii, zdot_eta, zdot_star, zdot_mu]

def zeq(out, eta_function, zone_width = 0.1, yocc = 0.00713, zin = 0, xi_out = 1, **kwargs):
    radii, gamma_eta, gamma_star, gamma_mu = gamma_coefficients(out, zone_width = zone_width, **kwargs)
    # gamma_tot = [a + b + c for a, b, c in zip(gamma_eta, gamma_star, gamma_mu)]
    _zeq = []
    for i in range(len(radii)):
        gamma_tot = gamma_eta[i] + gamma_star[i] + gamma_mu[i]
        x = yocc + zin * (gamma_tot - gamma_mu[i] - (xi_out - 1) * eta_function(radii[i]))
        x /= gamma_tot
        _zeq.append(x)
    return [radii, _zeq]

def boxcarsmoothtrend(xvals, yvals, window = 5):
    newyvals = len(yvals) * [0.]
    for i in range(len(xvals)):
        lower = max(i - window, 0)
        upper = min(i + window, len(xvals) - 1)
        newyvals[i] = np.mean(yvals[lower:upper])
    return newyvals
```
