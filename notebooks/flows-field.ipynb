# What happens in the presence of a velocity field?

If there is a velocity field, as opposed to a single uniform flow velocity, then we must take care to marginalize over the velocity field and come up with a proper estimate of how much gas will leave one ring in one timestep and how much will enter from neighboring rings. Since we do not have the full 6-D phase-space distribution in a multi-ring model, the best we can do is associate a particular $v_g$ with a distance $v_g \delta t$ in order to compute the area fractions $a$ for inward and outward flows, and then marginalize over the local velocity distribution $P(v_g | R)$. In general, it is simply understood here that one is choosing a ring width $\delta R$ and a timestep size $\delta t$ in their multi-ring model such that the velocity distribution $P(v_g | R)$ cuts off sufficiently at $v_g \ll \delta R / \delta t$, otherwise the model would be subject to numerical artifacts. However, in principle the velocity distribution extends to $\pm \infty$, so we retain this generality here under this caveat.

## The Inward Component

From the inward (i.e. $v_g < 0$) component of the local velocity distribution $P(v_g | R)$, there will be a *loss* to a ring's inner neighbor and a *gain* from its outer neighbor. The relevant area fractions are then given by

$$
\begin{align}
a_\text{in}(R) &=
\frac{-2 R v_g \delta t + v_g^2 \delta t^2}{2 R \delta R + \delta R^2}
\\
a_\text{in}(R + \delta R) &=
\frac{-2 (R + \delta R) v_g \delta t + v_g^2 \delta t^2}{2 (R + \delta R) \delta R + \delta R^2}
\end{align}
$$

The total rate of change in the mass of some element $x$ due to the radial flow can then be expressed as the difference between the source and sink terms as

$$
\begin{align}
\dot{M}_{x,\text{flow,in}} &=
Z_x(R + \delta R) M_g(R + \delta R) \int_{-\infty}^0
\frac{-2(R + \delta R) v_g + v_g^2 \delta t}{2(R + \delta R)\delta R + \delta R^2}
P(v_g | R + \delta R) dv_g -
Z_x(R) M_g(R) \int_{-\infty}^0
\frac{-2 R v_g + v_g^2 \delta t}{2 R \delta R + \delta R^2}
P(v_g | R) dv_g
\\
&= Z_x(R + \delta R) M_g(R + \delta R) \int_{-\infty}^0
\frac{-2(R + \delta R) v_g + v_g^2 \delta t}{2 R \delta R + 3 \delta R^2}
P(v_g | R + \delta R) dv_g -
Z_x(R) M_g(R) \int_{-\infty}^0
\frac{-2 R v_g + v_g^2 \delta t}{2 R \delta R + \delta R^2}
P(v_g | R) dv_g
\end{align}
$$

At this point, it is helpful to introduce taylor expansions for $Z_x(R)$, $M_g(R)$, and $P(v_g | R)$:

$$
\begin{align}
M_g(R + \delta R) &= M_g(R) + \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i
\\
Z_x(R + \delta R) &= Z_x(R) + \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i
\\
P(v_g | R + \delta R) &= P(v_g | R) + \sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i
\end{align}
$$

which yields the following next step

$$
\begin{align}
\dot{M}_{x,\text{flow,in}} &= \frac{Z_x(R) M_g(R)}{\delta R} \Bigg[
\\
&\quad
\left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2 R + 3\delta R}
\\
&\qquad
\int_{-\infty}^0 (-2Rv_g - 2\delta R v_g + v_g^2 \delta t) P(v_g | R)
\left(1 + \frac{1}{P(v_g | R)} \sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i\right)
dv_g
\\
&\quad - \frac{1}{2R + \delta R}
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t) P(v_g | R) dv_g \Bigg]
\\
&= \frac{Z_x(R) M_g(R)}{\delta R} \Bigg[
\\
&\quad
\left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2 R + 3\delta R}
\int_{-\infty}^0 (-2Rv_g -2\delta R v_g + v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad - \frac{1}{2R + \delta R}
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad + \left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R + 3 \delta R}
\int_0^\infty (-2Rv_g - 2\delta R v_g + v_g^2 \delta t)
\sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i dv_g
\Bigg]
\\
&= \frac{Z_x(R) M_g(R)}{\delta R} \Bigg[
\\
&\quad
\left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2 R + 3\delta R}
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad
- \frac{1}{2R + \delta R}
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad
+ \left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R + 3 \delta R}
\int_0^\infty (-2Rv_g + v_g^2 \delta t)
\sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i
\\
&\quad
- \left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{\delta R}{2R + 3\delta R}
\int_{-\infty}^0 2v_g P(v_g | R) dv_g
\\
&\quad
- \left(1 + \frac{1}{Z_x} \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 + \frac{1}{M_g} \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R + 3\delta R}
\int_{-\infty}^0 2v_g \sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^{i + 1} dv_g
\Bigg]
\end{align}
$$

At this point, we have to take the derivative with respect to $\delta R$ of the quantity in square brackets. This is required to apply L'Hopital's Rule as this approaches the $0/0$ indeterminate form in the limit that $\delta R \rightarrow 0$. Skipping intermediate steps for brevity:

$$
\begin{align}
\dot{M}_{x,\text{flow,in}} &= Z_x(R) M_g(R) \Bigg[
\left(\frac{\partial \ln Z_x}{\partial R}\frac{1}{2R} +
\frac{\partial \ln M_g}{\partial R}\frac{1}{2R} + \frac{-3}{4R^2}\right)
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t)P(v_g | R) dv_g
\\
&\quad + \frac{1}{4R^2} \int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t)P(v_g | R) dv_g
\\
&\quad + \frac{1}{2R} \int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t)
\frac{\partial P(v_g | R)}{\partial R} dv_g
\\
&\quad + \frac{1}{2R} \int_{-\infty}^0 2v_g P(v_g | R) dv_g
\\
&\quad - 0 \Bigg]
\\
&= Z_x(R) M_g(R) \Bigg[
\left(\frac{\partial \ln Z_x}{\partial R} \frac{1}{2R} + \frac{\partial \ln M_g}{\partial R}\frac{1}{2R}\right)
\int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t)P(v_g | R) dv_g
\\
&\quad + \left(\frac{-3}{4R^2} + \frac{1}{4R^2}\right) \int_{-\infty}^0 (-2Rv_g + v_g^2 \delta t)
P(v_g | R) dv_g
\\
&\quad + \frac{1}{2R} \int_{-\infty}^0 2 v_g P(v_g | R) dv_g
\\
&\quad + \frac{1}{2R} \int_{-\infty}^0 (-2R v_g + v_g^2 \delta t)
\frac{\partial P(v_g | R)}{\partial R} dv_g
\Bigg]
\\
&\rightarrow Z_x(R) M_g(R) \Bigg[
\left(-\frac{\partial \ln Z_x}{\partial R} - \frac{\partial \ln M_g}{\partial R}\right)
\int_{-\infty}^0 v_g P(v_g | R) dv_g
\\
&\quad - \frac{1}{R} \int_{-\infty}^0 v_g P(v_g | R) dv_g
\\
&\quad + \frac{1}{R} \int_{\infty}^0 v_g P(v_g | R) dv_g
\\
&\quad - \int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
\Bigg]
\end{align}
$$

where I have dropped the $v_g^2 \delta t$ term in the integrand in the final equality to apply the limit that $\delta t \rightarrow 0$.

$$
\begin{align}
\dot{M}_{x,\text{flow,in}} &= Z_x(R) M_g(R)
\left[\left(-\frac{\partial \ln Z_x}{\partial r} - \frac{\partial \ln M_g}{\partial R}\right)
\int_{-\infty}^0 v_g P(v_g | R) dv_g -
\int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g \right]
\\
&= -Z_x \dot{M}_\star \tau_\star
\left[\frac{\partial \ln M_g}{\partial R} + \frac{\partial \ln Z_x}{\partial R}
+ \frac{
    \int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_{-\infty}^0 v_g P(v_g | R) dv_g
}
\right] \int_{-\infty}^0 v_g P(v_g | R) dv_g
\\
&= -Z_x \dot{M}_\star \tau_\star
\left[\frac{1}{R} + \frac{\partial \ln \Sigma_g}{\partial R} + \frac{\partial \ln Z_x}{\partial R}
+ \frac{
    \int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_{-\infty}^0 v_g P(v_g | R) dv_g
}
\right] \int_{-\infty}^0 v_g P(v_g | R) dv_g
\end{align}
$$

## The Outward Component

From the outward component of the local velocity distribution, there will be a *gain* from a ring's inner neighbor and a *loss* to its outer neighbor. The relevant area fractions are then given by

$$
\begin{align}
a_\text{out}(R) &= \frac{2(R + \delta R)v_g \delta t - v_g^2 \delta t^2}{2R\delta R + \delta R^2}
\\
a_\text{out}(R - \delta R) &= \frac{2 R v_g \delta t - v_g^2 \delta t^2}{2(R - \delta R)\delta R + \delta R^2},
\end{align}
$$

yielding the following expression for the total rate of change in the metal mass due to the flow:

$$
\begin{align}
\dot{M}_{x,\text{flow,out}} &=
Z_x(R - \delta R)M_g(R - \delta R)
\int_0^\infty \frac{2R v_g - v_g^2\delta t}{2 (R - \delta R) \delta R + \delta R^2} P(v_g | R - \delta R) dv_g -
Z_x(R) M_g(R)
\int_0^\infty \frac{2(R + \delta R)v_g - v_g^2 \delta t}{2R \delta R + \delta R^2} P(v_g | R) dv_g
\\
&= Z_x(R - \delta R)M_g(R - \delta R)
\int_0^\infty \frac{2R v_g - v_g^2\delta t}{2R\delta R - \delta R^2} P(v_g | R - \delta R) dv_g -
Z_x(R) M_g(R)
\int_0^\infty \frac{2Rv_g + 2\delta Rv_g - v_g^2 \delta t}{2R \delta R + \delta R^2} P(v_g | R) dv_g
\end{align}
$$

Following the same procedure as above for the inward flow, we now substitute Taylor expansions in for $Z_x(R)$, $M_g(R)$, and $P(v_g | R)$ given by

$$
\begin{align}
Z_x(R - \delta R) &= Z_x(R) - \sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i
\\
M_g(R - \delta R) &= M_g(R) - \sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i
\\
P(v_g | R - \delta R) &= P(v_g | R) - \sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i
\end{align}
$$

Plugging in the Taylor expansions results in the following next step

$$
\begin{align}
\dot{M}_{x,\text{flow,out}} &=
\frac{Z_x(R) M_g(R)}{\delta R}\Bigg[
\\
&\quad
\left(1 - \frac{1}{Z_x}\sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 - \frac{1}{M_g}\sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R - \delta R} \int_0^\infty (2Rv_g - v_g^2 \delta t)
P(v_g | R)
\left(1 - \frac{1}{P(v_g | R)} \sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i\right) dv_g
\\
&\quad - \frac{1}{2R + \delta R} \int_0^\infty (2Rv_g + 2\delta R v_g - v_g^2 \delta t) P(v_g | R) dv_g
\Bigg]
\\
&= \frac{Z_x(R) M_g(R)}{\delta R}\Bigg[
\\
&\quad
\left(1 - \frac{1}{Z_x}\sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 - \frac{1}{M_g}\sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R - \delta R} \int_0^\infty (2Rv_g - v_g^2 \delta t)
P(v_g | R) dv_g
\\
&\quad
- \frac{1}{2R + \delta R}
\int_0^\infty (2Rv_g - v_g^2 \delta t)
P(v_g | R) dv_g
\\
&\quad
-\frac{\delta R}{2R + \delta R}
\int_0^\infty 2v_g P(v_g | R) dv_g
\\
&\quad
- \left(1 - \frac{1}{Z_x}\sum_{i = 1}^\infty \frac{\partial^i Z_x}{\partial R^i} \delta R^i\right)
\left(1 - \frac{1}{M_g}\sum_{i = 1}^\infty \frac{\partial^i M_g}{\partial R^i} \delta R^i\right)
\frac{1}{2R - \delta R} \int_0^\infty (2Rv_g - v_g^2 \delta t)
\sum_{i = 1}^\infty \frac{\partial^i P(v_g | R)}{\partial R^i} \delta R^i dv_g
\Bigg]
\end{align}
$$

Continuing to follow the procedure from above, we now take the derivative with respect to $\delta R$ of the quantity in square brackets in order to apply L'Hopital's Rule. Omitting the intermediate steps for brevity once more:

$$
\begin{align}
\dot{M}_{x,\text{flow,out}} &= Z_x(R) M_g(R) \Bigg[
\left(-\frac{\partial \ln Z_x}{\partial R}\frac{1}{2R} -
\frac{\partial \ln M_g}{\partial R}\frac{1}{2R} + 
\frac{1}{4R^2}\right)
\int_0^\infty (2Rv_g - v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad + \frac{1}{4R^2} \int_0^\infty (2Rv_g - v_g^2 \delta t) P(v_g | R) dv_g
\\
&\quad - \frac{1}{2R} \int_0^\infty 2v_g P(v_g | R) dv_g
\\
&\quad - \frac{1}{2R}\int_0^\infty (2Rv_g - v_g^2 \delta t) \frac{\partial P(v_g | R)}{\partial R} dv_g
\Bigg]
\\
&= Z_x(R) M_g(R) \Bigg[
\left(-\frac{\partial \ln Z_x}{\partial R} - \frac{\partial \ln M_g}{\partial R} + \frac{1}{2R}\right)
\int_0^\infty v_g P(v_g | R) dv_g
\\
&\quad
+ \frac{1}{2R} \int_0^\infty v_g P(v_g | R) dv_g - \frac{1}{R} \int_0^\infty v_g P(v_g | R) dv_g -
\int_0^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
\Bigg]
\\
&= -Z_x \dot{M}_\star \tau_\star \left[
\frac{\partial \ln Z_x}{\partial R} + \frac{\partial \ln M_g}{\partial R} +
\frac{
    \int_0^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_0^\infty v_g P(v_g | R) dv_g
}
\right] \int_0^\infty v_g P(v_g | R) dv_g
\\
&= -Z_x \dot{M}_\star \tau_\star \left[
\frac{1}{R} +
\frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln Z_x}{\partial R} +
\frac{
    \int_0^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_0^\infty v_g P(v_g | R) dv_g
}
\right] \int_0^\infty v_g P(v_g | R) dv_g
\end{align}
$$

## Combining Terms

$$
\begin{align}
\dot{M}_{x,\text{flow}} &= \dot{M}_{x,\text{flow,in}} + \dot{M}_{x,\text{flow,out}}
\\
&=
-Z_x \dot{M}_\star \tau_\star
\left[\frac{1}{R} + \frac{\partial \ln \Sigma_g}{\partial R} + \frac{\partial \ln Z_x}{\partial R}
+ \frac{
    \int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_{-\infty}^0 v_g P(v_g | R) dv_g
}
\right] \int_{-\infty}^0 v_g P(v_g | R) dv_g
\\
&\quad
-Z_x \dot{M}_\star \tau_\star \left[
\frac{1}{R} +
\frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln Z_x}{\partial R} +
\frac{
    \int_0^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_0^\infty v_g P(v_g | R) dv_g
}
\right] \int_0^\infty v_g P(v_g | R) dv_g
\\
&=
-Z_x \dot{M}_\star \tau_\star
\left[
\left(\frac{1}{R} +
\frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln Z_x}{\partial R}
\right) \int_{-\infty}^0 v_g P(v_g | R) dv_g +
\int_{-\infty}^0 v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
\right]
\\
&\quad
-Z_x \dot{M}_\star \tau_\star \left[
\left(\frac{1}{R} +
\frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln Z_x}{\partial R}
\right) \int_0^\infty v_g P(v_g | R) dv_g +
\int_0^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
\right]
\\
&= -Z_x \dot{M}_\star \tau_\star \left[
\frac{1}{R} + \frac{\partial \ln \Sigma_g}{\partial R} + \frac{\partial \ln Z_x}{\partial R} +
\frac{
    \int_{-\infty}^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
}{
    \int_{-\infty}^\infty v_g P(v_g | R) dv_g
}
\right] \int_{-\infty}^\infty v_g P(v_g | R) dv_g
\end{align}
$$

By substituting in for the mean gas velocity and applying one of the fundamental theorems of calculus, we can simplify this expression considerably. With

$$
\int_{-\infty}^\infty v_g \frac{\partial P(v_g | R)}{\partial R} dv_g
= \frac{\partial}{\partial R} \int_{-\infty}^\infty v_g P(v_g | R) dv_g
= \frac{\partial \langle v_g \rangle}{\partial R}
$$

and

$$
\langle v_g \rangle = \int_{-\infty}^\infty v_g P(v_g | R) dv_g,
$$

we arrive at the following expression for the metal flow rate

$$
\dot{M}_{x,\text{flow}} = -Z_x \dot{M}_\star \tau_\star \langle v_g \rangle
\left[
\frac{1}{R} +
\frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln Z_x}{\partial R} +
\frac{\partial \ln \langle v_g \rangle}{\partial R}
\right]
$$

## Generalized Flow Coefficients

We now arrive at the following expressions for $\mu_\text{flow}$ and $\gamma_\text{flow}$, generalized to the case where there is a velocity distribution $P(v_g | R)$ at each radius $R$.

$$
\begin{align}
\gamma_\text{flow} &\equiv -\tau_\star \langle v_g \rangle
\left[\frac{1}{R} + \frac{\partial \ln \Sigma_g}{\partial R} +
\frac{\partial \ln \langle v_g \rangle}{\partial R}
\right]
\\
\mu_\text{flow} &\equiv \gamma_\text{flow} -
\tau_\star \langle v_g \rangle \frac{\partial \ln Z}{\partial R}.
\end{align}
$$

## Implications

These expressions have a number of interesting implications. First and foremost, they suggest that under the assumption of vertical and azimuthal symmetry (which one could essentially call a "1-D Galactic disk model"), the *entire* effect of the radial gas flow on both the gas and metal mass budgets, and by extension metallicity, is specified by the mean. Information on the shape of the velocity distribution is not required.

One possible test of this prediction with our multi-zone models, which would be quite straightforward, is to introduce a flow with both inward and outward gas exchange. The relevant area fractions would be given by

$$
\begin{align}
a_\text{in}(R) &= f_\text{in} \frac{2Rv_{g,\text{in}} \delta t + v_{g,\text{in}}^2 \delta t^2}{2R\delta R + \delta R^2}
\\
a_\text{out}(R) &= \left(1 - f_\text{in}\right) \frac{2(R + \delta R)v_{g,\text{out}} \delta t + v_{g,\text{out}}^2 \delta t^2}{2R\delta R + \delta R^2},
\end{align}
$$

where $f_\text{in}$ and $f_\text{out} \equiv 1 - f_\text{in}$ are the fractions of the ISM mass that is flowing in and out, respectively, and $v_{g,\text{in}}$ and $v_{g,\text{out}}$ are the corresponding velocities. The mean velocity in this case would be given by

$$
\langle v_g \rangle = f_\text{in} v_{g,\text{in}} + \left(1 - f_\text{in}\right) v_{g,\text{out}}.
$$

A comparison of one model with a single, inward flow velocity and another with both inward and outward components satisfying the above expression would be able to test this hypothesis. Based on the analytic arguments, I would expect the results to be equivalent within numerical noise.
